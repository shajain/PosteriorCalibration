```python
from MonotonicPosterior.generateData import generateData
from misc import sortedplot as sp
import torch
# X,y is generated with data prior
X,y, post1, dg = generateData(n=200, mu0=-1, sig0=2, alpha=0.5)
# alpha below is the true prior
alpha = 0.3
postTrue = dg.pn_posterior(X, a=alpha)
sp.hist(X[(y == 1).flatten(), :], bins=50, density=True, alpha=0.5)
sp.hist(X[(y == 0).flatten(), :], bins=50, density=True, alpha=0.5)
sp.sortedplot(X, postTrue, label='True Posterior')
sp.sortedplot(X, post1, label = 'Data Posterior')
sp.legend()
sp.show()
```

```python
from MonotonicPosterior.computePosterior import computePosteriorFromEnsemble
modelAvg_func, modelMedian_func, modelRobustMean_func, Ensemble = computePosteriorFromEnsemble(X,y, alpha=alpha, test_size=0.2, num_ensemble=25, epochs=1000)
sp.sortedplot(X,postTrue, label = "True")
X_torch=torch.tensor(X,dtype=torch.float)

postEsts = [model(X_torch).detach().numpy() for model in Ensemble]
for (i,postEst) in enumerate(postEsts):
    sp.sortedplot(X, postEst, label="post " + str(i))
sp.legend()
sp.show()


sp.sortedplot(X,postTrue, label = "True Posterior")
postAvg = modelAvg_func(X_torch)
postMedian = modelMedian_func(X_torch)
postRobustAvg = modelRobustMean_func(X_torch)
sp.sortedplot(X, postAvg, label="Avg Est.")
sp.sortedplot(X, postMedian, label="Median Est.")
sp.sortedplot(X, postRobustAvg, label="Robust Avg Est.")
sp.legend()
sp.show()
```
